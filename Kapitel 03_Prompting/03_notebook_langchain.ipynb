```python
URL = "http://132.199.138.16:27090/v1"
```

```python
from langchain_openai import ChatOpenAI
from pydantic import BaseModel, Field
from typing import Optional
```

```python
llm = ChatOpenAI(
    model="google/gemma-3-4b-it",
    openai_api_key="token-abc123",
    openai_api_base=URL,
    temperature=0,
    model_kwargs={"stop": ["\n"]},
    
)
```

```python
def get_weather(city_name: str) -> int:
    """ Get the current weather for a city.

    Args:
        city_name (str): The name of the city to get the weather for.
    """
    return 25  # Dummy temperature value for example purposes


class Response(BaseModel):
    """Response that may or may not use weather data."""
    
    # Optionale Tool-Parameter
    city_name: Optional[str] = Field(None, description="City name if weather is requested")


tools = [get_weather]

llm_with_tools = llm.bind_tools(tools, tool_choice="auto")# .with_structured_output(Response)
```

```python
query = "Wie ist es dir"
result = llm_with_tools.invoke(query)
```
