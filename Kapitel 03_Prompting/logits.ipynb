```python
import numpy as np
import matplotlib.pyplot as plt

# Logits für 5 Tokens
logits = np.array([4.3, 2.3, 1.4, 0.9, 0.2])
labels = ['sig1', 'sig2', 'sig3', 'sig4', 'sig5']

# Temperature-Werte, die wir testen wollen
temperatures = [k/100 for k in range(0, 2000)]

# Softmax-Funktion mit Temperature
def softmax(x, temperature=1.0):
    x = x / temperature
    e_x = np.exp(x - np.max(x))  # für numerische Stabilität
    return e_x / e_x.sum()

# Plot vorbereiten
plt.figure(figsize=(10, 6))

# Für jedes Token: Wahrscheinlichkeiten über alle Temperaturen berechnen und plotten
for i, label in enumerate(labels):
    token_probs = []
    for T in temperatures:
        probs = softmax(logits, temperature=T)
        token_probs.append(probs[i])
    plt.plot(temperatures, token_probs, label=f'{label} ({logits[i]})')

# Plot-Details
plt.title('Einfluss der Temperature auf Softmax-Wahrscheinlichkeiten')
plt.xlabel('Temperature')
plt.ylabel('Wahrscheinlichkeit')
plt.ylim(0, 1)
plt.legend(title='Token (Logit)')
plt.grid(True)
plt.tight_layout()
plt.show()
```
